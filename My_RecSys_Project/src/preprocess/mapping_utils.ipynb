```python
import pandas as pd

test_df = pd.DataFrame({
    'user': [11, 13 ,15 ,17],
    'item': [3,5,8,11]
})
```

```python
def build_mappings(values,start_idx, add_pad=False):

    uniques = values.sort_values().unique()

    # add pad일 경우 아닐경우 매핑
    if add_pad:
        idx2value = {v: i + start_idx for i, v in enumerate(uniques)}
        value2idx = {v: i + start_idx for i, v in enumerate(uniques)}
    else:
        idx2value = {v: i + start_idx for i, v in enumerate(uniques)}
        value2idx = {v: i + start_idx for i, v in enumerate(uniques)}
    
    return idx2value, value2idx

user2idx, idx2user = build_mappings(test_df['user'], 0, add_pad=True)

item2idx, idx2item = build_mappings(test_df['item'],0,add_pad=False)

print(type(user2idx))
```
